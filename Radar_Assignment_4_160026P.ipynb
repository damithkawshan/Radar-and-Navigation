#Q2 : Derivation of Received signals by the target

Two isotropic antenna elements ($A_1$ and $A_2$) are rotating with a frequency of $f_m$ around the center antenna element $A_0$ in a circle of radius $R$. (As shown in $Fig (2.1)$)

$A_1$ transmits a signal with frequency of $f_c+f_0$.

$A_2$ transmits a signal with frequency of $f_c-f_0$.

$A_0$ transmits a signal with frequency of $f_c$ and it's amplitude modulated by message signal $\sin(2\pi f_mt)$.

###Derivation

Assumptions made:
*   $|A(t)|\:= 1$ for signals $A_1$ and $A_2$
*   $d \gg R$ 
*   for $1 \gg |x|$ we can assume that $\sqrt{1-x} \approx 1-\frac{x}{2}$ and $\sqrt{1+x} \approx 1+\frac{x}{2}$

From cosine rule,

\begin{align}
d_1 &= \sqrt{R^2+d^2-2dR\cos(2\pi f_mt-\theta)} \tag{2.1}\\
&= d\sqrt{1+\frac{R^2}{d^2}-2\frac{R}{d}\cos(2\pi f_mt-\theta)}\\
&= d\sqrt{1-2\frac{R}{d}\cos(2\pi f_mt-\theta)} \because d \gg R \implies \frac{R^2}{d^2} \approx 0\\
d_1&=d-R\cos(2\pi f_mt-\theta) \because \sqrt{1-x} \approx 1-\frac{x}{2} \tag{2.2}
\\
\text{similarly we can obtain}
\\
d_2&=d+R\cos(2\pi f_mt-\theta) \tag{2.3}
\end{align}

According to derived equation in $(2.1)$ received signals from $A_1$ and $A_2$ can be written as follows:

\begin{align}
V_{A1}&=\sin(2\pi (f_c+f_0)t+\phi_1)\\
&=\sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda} d_1)\\
&=\sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda}(d-R\cos(2\pi f_mt-\theta))) \tag{2.4}\\
\text{similarly}\\
V_{A2}&=\sin(2\pi (f_c-f_0)t+\frac{2\pi}{\lambda}(d+R\cos(2\pi f_mt-\theta))) \tag{2.5}\\
\therefore \text{Signal received by the target} (V_{Rx}^0)\\
V_{Rx}^0&= V_{A1} + V_{A2}\\
V_{Rx}^0&= \sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda}(d-R\cos(2\pi f_mt-\theta))) \\&+\sin(2\pi (f_c-f_0)t+\frac{2\pi}{\lambda}(d+R\cos(2\pi f_mt-\theta))) \tag{2.6}\\
\end{align}

### Parameters of the hypothetical system

###Received signal by the Target

Received signal by the target emitted from $A_0,A_1,A_2$ Antenna elements.

\begin{align}
V_{A0}&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct - \frac{2\pi d}{\lambda}\right) \tag{3.1}
\end{align}

Combine with $(2.6)$ we can obtain composite signal received by the target.

\begin{align}
V_{R_x}&=V_{A0}+V_{A1}+V_{A2}\\
V_{R_x}&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct - \frac{2\pi d}{\lambda}\right)+\sin(2\pi (f_c+f_0)t+\frac{2\pi}{\lambda}(d-R\cos(2\pi f_mt-\theta))) \\&+\sin(2\pi (f_c-f_0)t+\frac{2\pi}{\lambda}(d+R\cos(2\pi f_mt-\theta))) \tag{3.2}
\end{align}

If we take $V_{A0}$ as our reference signal relative phase shift can be obtained as follows:
\begin{align}
\Delta\phi_1&= \frac{2\pi}{\lambda}(d-d_1) &=\frac{2\pi}{\lambda}R\cos(2\pi f_mt-\theta) &=\Delta\phi \tag{3.3}\\
\Delta\phi_2&= \frac{2\pi}{\lambda}(d2-d)&=-\frac{2\pi}{\lambda}R\cos(2\pi f_mt-\theta) &=-\Delta\phi \tag{3.4}\\
\end{align}

We can rewrite $(3.2)$ as:
\begin{align}
V_{R_x}&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct \right)+\sin(2\pi (f_c+f_0)t+\Delta\phi) \\&+\sin(2\pi (f_c-f_0)t-\Delta\phi)\\
&=(1-M\sin(2\pi f_mt))\sin\left(2\pi f_ct \right)+2\sin(2\pi f_ct)\cos(2\pi f_mt +\Delta\phi )\\
&=(1-M\sin(2\pi f_mt)+2\cos(2\pi f_mt +\Delta\phi ))\sin\left(2\pi f_ct \right)\\
V_{R_x}&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_mt +\frac{2\pi R}{\lambda}\cos(2\pi f_mt-\theta) \right)\right\}\sin\left(2\pi f_ct \right)\tag{3.5}
\end{align}



```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; 
#sns.set_theme(palette="flare")
# plt.style.use("dark_background")

from scipy.signal import freqz,hilbert
import sys

#Define Parameters
#-------------------------------------------------------------------------------
f_c = 1e6 # Carrier Frequency 
f_0 = 960 # Freq offset
f_m = 30 # Rotation frequency of VOR system
d = 2*1e5 # Distance to the target
theta = ((2*math.pi)/360)*30 # Phase angle from magnetic north 
C = 3*1e8 # Speed of Light
M = 0.3 # Modulation index
delta = 16 

#-------------------------------------------------------------------------------
```

```python
#Initial Calculations
lmbda = C / f_c # Wavelength
R = 8*lmbda/math.pi # Radius of the VOR antenna 

w_m = 2*math.pi*f_m
w_0 = 2*math.pi*f_0
w_c = 2*math.pi*f_c

freq_div = delta*f_m
E = 1.0/M

#sampling frequency
F_s = int(10*f_c)

#time vector
t   = np.linspace(0, 1, F_s+1)
phase_shift = (2*math.pi*d)/lmbda

print('Radius of the VOR antenna circle (R):',str(R),'m')
print('----------------------------------------------------------')
print('Phase shift due to distance: ',str(phase_shift),'rad')
```

```python
d_1 = d - R*np.cos(w_m*t-theta)
d_2 = d + R*np.cos(w_m*t-theta)

delt_phi_1 = (2*math.pi/lmbda)*(d_1)
delt_phi_2 = (2*math.pi/lmbda)*(d_2)

V_A1 = E*np.sin((w_c+w_0)*t - delt_phi_1)
V_A2 = E*np.sin((w_c-w_0)*t - delt_phi_2)

V_A0 = E*(1-M*np.sin(w_m*t))*np.sin(w_c*t-phase_shift)

plt.figure(figsize=(25,5))
plt.plot(t[0:3000], V_A1[0:3000])
plt.ylim((-4 ,4))
plt.title('Recieved Signal from Antenna Element A1')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()

plt.figure(figsize=(25,5))
plt.plot(t[0:3000], V_A2[0:3000])
plt.ylim((-4 ,4))
plt.title('Recieved Signal from Antenna Element A2')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()

plt.figure(figsize=(25,5))
plt.plot(t[:300000], V_A0[:300000])
plt.title('Recieved Signal from Antenna Element A0')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()
```

```python
V_Rx = V_A0+V_A1+V_A2 #Rx signal

plt.figure(figsize=(50,10))
plt.plot(t[0:1000000], V_Rx[0:1000000])
plt.title('Recieved composite signal $V_{Rx}$ emitted from VOR antenna elements from the receiver of the target',fontsize='x-large')
plt.xlabel('Time(s)')
plt.ylabel('Magnitude')
plt.show()
```

```python
#supporting functions

def plot_FFT(abs_FFT,F_s,range,name):

  N = len(abs_FFT)
  freq_axis = np.fft.fftfreq(N, 1/F_s)
  plt.figure(figsize=(20, 5))
  ax=plt.ylabel('$|V(f)|^2$')
  ax=plt.xlabel('Frequency (Hz)')
  ax=plt.title('Frequency Response of :  '+name)
  plt.xlim(range)
  ax=plt.plot(freq_axis,  (2/N)*abs_FFT)

  return None

def LPF(cutoff,fs,N,plot,name):
  H = np.zeros((1,N))
  H[0,0:cutoff+1] = np.ones((1,cutoff+1))
  H[0,-cutoff:] = np.ones((1,cutoff))
  H=H[0]
  print(H.shape)

  if plot:
    plt.figure(figsize=(20,4))
    ax=plt.ylabel('$|H(f)|^2$')
    ax=plt.xlabel('Frequency (Hz)')
    ax=plt.title('Lowpass filter for : '+name)
    plt.plot(H,color='r')
    plt.xlim((0,cutoff+50))
  return H


def BPF(cutoff_L,cutoff_R,fs,N,plot,name):
  L_idx = cutoff_L
  R_idx = cutoff_R

  H = np.zeros((1,N))
  print(H.shape)
  H[0,cutoff_L:cutoff_R+1 ]= np.ones((1,cutoff_R-cutoff_L+1))
  H[0,-cutoff_R-1:-cutoff_L]=np.ones((1,cutoff_R-cutoff_L+1))
  H = H[0]
  print(H.shape)

  if plot:
    plt.figure(figsize=(20,4))
    ax=plt.ylabel('$|H(f)|^2$')
    ax=plt.xlabel('Frequency (Hz)')
    ax=plt.title('Banpass filter for : '+name)
    plt.plot(H,color='r')
    plt.xlim((0,cutoff_R+50))
  return H   
```

###Stage 01: Remove carrier signal

To remove the carrier and obtain baseband signal, product detector followed by a low pass filter is used as in conventional AM demodulators.

This yields to 

\begin{align}
V_{R_x}&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_mt +\frac{2\pi R}{\lambda}\cos(2\pi f_mt-\theta) \right)\right\}\sin\left(2\pi f_ct \right) \times 2\sin(2\pi f_ct)\\
&=\left\{1-M\sin(2\pi f_mt)+2\cos\left(2\pi f_mt +\frac{2\pi R}{\lambda}\cos(2\pi f_mt-\theta) \right)\right\}(1-\cos(4\pi f_ct))
\end{align}

We can obtain the baseband signal by applying a low pass filter with a cutoff frequency of 10kHz.

```python
#Frequency Domain Response
Rx_FFT=np.fft.fft(V_Rx)
abs_Rx_FFT = (np.abs(Rx_FFT)**1).flatten()

plot_FFT(abs_Rx_FFT,F_s,(1e6-2000,1e6+2000),'$V_{Rx}$ around carrier frequency ($1 MHz$)')
```

```python
V_mod = V_Rx*(2*np.sin(w_c*t-phase_shift)) #suppress carrier

#Frequency Domain Response
mod_FFT=np.fft.fft(V_mod)
mod_abs_FFT = (np.abs(mod_FFT)**1).flatten()

#low pass filter to obtain baseband signal
H0_LPF=LPF(10000,F_s,len(V_mod),0,'filter product detector output')
plot_FFT(mod_abs_FFT,F_s,(-2.7e6,2.7e6),'Output of the product detector $V_{Rx}$')
plt.plot(freq_axis,H0_LPF,'r')
plt.show()

W_mod = mod_FFT*H0_LPF
W_mod[0] = 0 #remove dc component

V_demod = np.real(np.fft.ifft(W_mod))

plot_FFT((np.abs(W_mod)**1).flatten(),F_s,(-2000,2000),'LPF filtered AM demodulated $V_{Rx}$')                            

fig= plt.figure(figsize=(50,10))
plt.plot(t, V_demod)
plt.title('Stage 01: Demodulated Signal')
plt.xlabel('Time (s)') 
plt.ylabel('Magnitude')
plt.xlim((0, 0.08))
plt.show()
```

###Stage 04: FM demodulation of variable 30Hz signal

FM demodulation has been performed in order to extract variable 30Hz signal.

```python
def fm_discriminator(x, freq_dev, f_c,t=t):

    R_x = hilbert(x)*np.exp(-1j*2*np.pi*f_c*t)

    phi = np.arctan2(np.imag(R_x), np.real(R_x))
    ret = np.diff(np.unwrap(phi)/(2*math.pi*freq_dev))*t.shape[0]

    return ret
```

```python
fitered_var=fm_discriminator(np.real(filt_var_ref), freq_dev=30*16,f_c=f_0)
fitered_var=(fitered_var-np.mean(fitered_var))
```

```python
#phase extraction

ref_fft = np.fft.fft(filtered_ref - np.mean(filtered_ref))
var_fft = np.fft.fft(fitered_var- np.mean(fitered_var))

#get the w_m frequency bin
var_fft_bin= np.argmax(abs(var_fft))
ref_fft_bin = np.argmax(abs(ref_fft))

var_phase=np.angle(var_fft[var_fft_bin])
ref_phase=np.angle(ref_fft[ref_fft_bin])

phase_diff=np.abs(var_phase-ref_phase)

print('----------------------------')
print('phase difference : ',np.round(phase_diff*180/math.pi,2),'deg')
print('----------------------------')
```

#Q5 : Simulate the Rotation Electronically

Fixed number of omni directional antennas are arranged in the circle with the radius $R$ of previous setup. Rotation can be simulated by activating/deactivating radial pairs of the antenna setup.

Since radiation cannot be switched abruptly, zero switching concept has been introduced. cosine blending function is used to ramping up/down the power.

We can divide antenna pairs into odd and even groups and handing over will be done from odd-to-odd and even-to-even.

\begin{align}
f_{blend}&=\frac{2\pi f_m}{4\times(2\pi / \text{antenna count})}\\
&=\frac{2\pi \times 30}{4\times(2\pi /48)}\\
&=360Hz
\end{align}

###Blending odd and even function generation

\begin{align}
\text{cosine blending function for odd antennas}&= |\sin(2\pi f_{blend}t)|\\
\text{cosine blending function for even antennas}&= |\cos(2\pi f_{blend}t)|
\end{align}

```python
antennas=48
blend_freq= (np.pi*f_m)/(2*(2*np.pi / antennas))
pulse_width= 2/(f_m*antennas)
print("cosine blending frequency: ", blend_freq)
```

```python
halfw_sampled=int(0.5*pulse_width*F_s)+1
sample_period=int(F_s/f_m)+1
rep=31
beta=360/antennas

t_arr=np.arange(0,2/f_m,1/F_s)
```

```python
discr_sin=np.sin(blend_freq*t_arr)
discr_sin[2*halfw_sampled-1:sample_period]=0
odd_func=np.array(list(discr_sin[:sample_period])*rep)

discr_cos=np.cos(blend_freq*t_arr)
discr_cos[halfw_sampled-1:sample_period-halfw_sampled]=0
even_func=np.array(list(discr_cos[:sample_period])*rep)
```

##Generate the composite received signal with fixed antenna array

```python
#simulate rotation

V_Rx=0
for i in range(antennas):
  if i==0:
    angle=beta/2
    pulse=even_func
  else:
    angle+=beta
    pulse=np.array(list(np.zeros(((i-1)*halfw_sampled)))+list(odd_func))
  blend=pulse[0:N]
  dx=R*np.cos((theta-angle)*np.pi/180)
  phi_1=2*np.pi*(d-dx)/lmbda
  phi_2=2*np.pi*(d+dx)/lmbda

  upper_sp=E*np.sin(2*np.pi*(f_c+f_0)*t-phi_1)*blend
  lower_sp=E*np.sin(2*np.pi*(f_c-f_0)*t-phi_2)*blend
  V_Rx+=upper_sp+lower_sp
```
